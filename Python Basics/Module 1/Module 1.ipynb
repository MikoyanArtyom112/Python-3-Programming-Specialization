## Updating Variables

In variable assignment and reassignment we assign values to a variable and python creates a reference in a variable's value. Additionally, when we perofrm an operation, Python first evaluates the ride side then updates the new value. Furthermore incrementing and decrementing is also possible and we may reassign variables multiple times. The following cell below demonstrates this process.

```python
x = 6
print(x)

x = x + 1
print(x)
```

The variable x becomes reassigned into a new value. This is how Python reassignment works when we use the same variable.

```python
x = 0
print(x)

x += 1 # increment by x; same as x = x + 1
print(x)

x -= 1 # decrement by x; same as x = x - 1
print(x)
```
