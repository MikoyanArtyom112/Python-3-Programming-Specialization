## First Turtle Program

Here we try to create a turtle on the center of teh screen that can move with basic commands. Furthermore it is an instance of the class turle and methods like the commands forward and left are introduced to give it some mobility We also set attributes and shape for our Alex (the turtle).

```python
import turtle
```

```python
wn = turtle.Screen() # creates a graphic window
```

```python
alex = turtle.Turtle() # creates a turtle named alex
```

```python
# we tell alex to move around
alex.forward(150)
alex.left(90)
alex.forward(75)
```

| Term       | Description                                           | Example                                |
|------------|-------------------------------------------------------|----------------------------------------|
| Class      | A blueprint for creating objects                      | `turtle.Turtle`                        |
| Object     | An instance of a class (the actual thing)             | `alex = turtle.Turtle()`               |
| Attribute  | A variable stored inside an object (describes state) | `alex.color`, `alex.position()`       |
| Method     | A function inside an object (defines behavior)       | `alex.forward(100)`, `alex.left(90)`  |

| Term       | Think of it as...           | What it does                          | Example (Turtle)                  |
|------------|-----------------------------|----------------------------------------|-----------------------------------|
| Class      | A recipe or blueprint        | Describes how something should work    | `turtle.Turtle`                  |
| Object     | The thing made from a class  | A working version of the class         | `alex = turtle.Turtle()`         |
| Attribute  | A fact or detail about it    | Stores information like color or size  | `alex.color()` (gets the color)  |
| Method     | An action it can do          | Makes the object do something          | `alex.forward(100)`              |
